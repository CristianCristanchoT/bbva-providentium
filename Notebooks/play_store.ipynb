```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from tqdm import tqdm 
from tqdm import trange
import time
from google_play_scraper import app, Sort, reviews, reviews_all
from pprint import pprint
import json
import re
import pprint
from datetime import datetime, timedelta
```

```python
bbva_colombia = 'co.com.bbva.mb'
bbva_mexico = 'com.bancomer.mbanking'
bbva_espa침a = 'com.bbva.bbvacontigo'
```

```python
def obtener_ultimos_comentarios_play_store(id_app, horas_atras = 1):
    
    try:
    
        fecha_target = datetime.now() - timedelta(hours=int(horas_atras), minutes=0)

        results = []

        result, continuation_token = reviews(
            id_app,
            lang='es', # defaults to 'en'
            country='us', # defaults to 'us'
            sort=Sort.NEWEST, 
            count=1, # defaults to 100
        )

        if result[0]['at'] > fecha_target:

            results.append(result[0])

            while True:

                result, continuation_token = reviews(
                    id_app,
                    continuation_token=continuation_token # defaults to None(load from the beginning)
                )

                if result[0]['at'] > fecha_target:

                    results.append(result[0])

                else:

                    break

        df = pd.json_normalize(results)
        
        return df
    
    except Exception as e:
        
        return e
    
```

```python
comentrarios_bbva_colombia = obtener_ultimos_comentarios_play_store(bbva_colombia, 240)
comentrarios_bbva_colombia.to_csv('app_bbva_colombia.csv')
comentrarios_bbva_colombia
```
