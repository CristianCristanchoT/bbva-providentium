```python
from sentiment_analysis_spanish import sentiment_analysis

sentiment = sentiment_analysis.SentimentAnalysisSpanish()
```

```python
def analysis_de_sentimiento(frase):
    
    sentimiento_numerico = sentiment.sentiment(frase)
    
    if sentimiento_numerico > 0.5:
        
        sentimiento = 'Positivo'
    
    else:
        
        sentimiento = 'Negativo'
        
    return sentimiento, sentimiento_numerico
```

```python
analysis_de_sentimiento('Mañana es quincena y se cae bancolombia')
```

```python
sentimiento = []
sentimiento_numerico = []

for index, row in tqdm(data.iterrows()):
    
    resultado = analysis_de_sentimiento(row['Texto'])
    
    sentimiento.append(resultado[0])
    sentimiento_numerico.append(round(resultado[1], 2))

data['sentimiento'] = sentimiento
data['sentimiento_numerico'] = sentimiento_numerico 
```
